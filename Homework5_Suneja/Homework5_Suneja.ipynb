## Problem 1: Basic Plotting with Matplotlib (1pt)

 Objective: create a line plot using Matplotlib to visualize the trend of a given dataset

* Dataset: use np.linspace(0,10,100) for the x-values and np.sin(x) for the y-values
* Add title to plot
* x-axis = 'X'
* y-axis = 'sin(X)'

```python
import numpy as np
import matplotlib.pyplot as plt

# creating the dataset as per instructions
x = np.linspace(0,10,100)
y = np.sin(x)

# creating the plot
plt.plot(x,y, linestyle = '--', color = 'blue')
plt.title("Sine Function")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.show()
```

## Problem 2: Exploring Distributions with Seaborn (1 pt)

Objective: Use Seaborn to visualize the distribution of the total_bill column from the tips dataset

* Task 1: Create a histogram using sns.histplot
* Task 2: Overlay a Kernal Density Estimate (KDE) on the histogram
* Requirements: (1) set the number of bins to 20 for the histogram and (2) add appropriate titles for both plots


```python
import seaborn as sns
import matplotlib.pyplot as plt

# dataset: tips
tips = sns.load_dataset("tips")
print(tips.head())

# Task 1: create a histogram using sns.histplot
plt.figure(figsize=(8,5))
sns.histplot(tips["total_bill"], bins = 20, kde = False, color = 'blue')
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show

# Task 2: Create a histogram with KDE overlay
# Task 2: Create a histogram with KDE overlay
plt.figure(figsize=(8, 5))
sns.histplot(tips["total_bill"], bins=20, kde=True, color='blue')
plt.title("Histogram with KDE Overlay of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Kernal Density Estimate")
plt.show()
```

## Problem 3: Categorical Data Analysis (2 pts)

Objective: Compare the distribution of tips received by time of day (lunch vs. dinner) using box plots and violin plots.
* Dataset: tips dataset available in seaborn
* Requirements: (1) add a legend to distinguish between smokers and non-smokers and (2) include a regression line for each group (hint: use sns.lmplot)


```python
import seaborn as sns
import matplotlib.pyplot as plt

# dataset: tips
tips = sns.load_dataset("tips")

# setting up a blank figure that has both plots (2 subplots) in it
sns.set_style("whitegrid")
fig,axes = plt.subplots(1,2,figsize = (15,8))

# box plot of tips by time of day
sns.boxplot(x = "time", y = "tip", hue = "smoker", data = tips, ax = axes[0], palette = "Set2") # main box plot command
axes[0].set_title("Box Plot of Resturaunt Tips by Time of Day") # telling it what the title is
axes[1].legend(title = "Smoker")  # telling it to include a legend

# violin plot of tips by time of day
sns.violinplot(x = "time", y = "tip", hue = "smoker", data = tips, ax = axes[1], split = True, palette = "Set2")  # similar to box plot, but telling it to be split
axes[1].set_title("Violin Plot of Resturaunt Tips by Time of Day")
axes[1].legend(title = "Smoker")

# printing the plots
plt.tight_layout()  # make it a neater graph
plt.show()
```

## Problem 5: Heatmap of Correlation Matrix (2 pts)
Objective: visualize the correlation matrix of the tips dataset using a heatmap
* Task: calculate the correlation matrix and visualize it using sns.heatmap
* Requirements: (1) annotate each cell in the heatmap with the correlation coefficient and (2) use a diverging color palette to highlight differences in correlation values


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# dataset: tips
tips = sns.load_dataset("tips")

# making correlation matrix (numerical columns)
corr_matrix = tips.corr(numeric_only = True)
plt.figure(figsize = (10, 8))

# creating the heatmap
sns.heatmap(corr_matrix, annot = True, fmt = ".2f", cmap = "coolwarm", vmin = 0, vmax = 1, linewidths = 0.5)
plt.title("Correlation Maxtrix of Tips Dataset")
plt.show()
```

## Problem 6: Pair Plot Analysis (2 pts)
Objective: use seaborn's pairplot function to visualize relationships between all numerical variables in the tips dataset
